```python
import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('01.csv')
month = ["Jan", "Feb", "March", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y=len(month)

for x in range(y):
    country_data = df["List"]
    medal_data = df[month[x]]
    colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b", "#8d764b"]
    explode = (0, 0, 0, 0, 0, 0)  
    plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
    autopct='%1.1f%%', shadow=True, startangle=140)
    plt.title(month[x])
    plt.show()
```
